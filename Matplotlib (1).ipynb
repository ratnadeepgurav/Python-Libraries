# Python Matplotlib Library

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,4]
y = [3,7,5,4]
color_graph = ["red",'yellow',"green",'pink']

# Diplay the Linear Graph
plt.plot(x,y)

plt.show()

#Diplay the Bar Graph
plt.bar(x,y,color = color_graph)

plt.show()
```

# Bar Plot

```python
x = ["Python" , "C" ,"CPP", "Java"]
y = [3,7,5,4]
color_plot = ['y','g','r','b']

# Labeling to the plot 
plt.xlabel("Programming Language", fontsize = 10)
plt.ylabel("Popular", fontsize = 10)
plt.title("Treading Language",fontsize = 14)


# Diplay the Bar Plot
plt.bar(x,y,color = color_plot, width = 0.7 , align = 'edge' , edgecolor ='black' , linewidth = 1.7, linestyle = ':', alpha = 0.5,label= x)
#Align property for adjust postion of x axis name center to left edge or vice versa
#edgecolor :- color to boundry bar lines
#linewidth :- width of bar lines 
#linestyle :- type of bar lines (:,solid , ) 
#alpha :- opacity of bar graph color

plt.legend() # when label property are need to apply , lenged are used to display this content on bar chart 
plt.show()
```

```python
x = ["Python" , "C" ,"CPP", "Java"]
y = [3,7,5,4]
z = [2,3,3,2]

# Labeling to the plot 
plt.xlabel("Programming Language", fontsize = 10)
plt.ylabel("Popular", fontsize = 10)
plt.title("Treading Language",fontsize = 14)


# Diplay the Bar Plot
plt.bar(x,y, width = 0.7, color = 'pink', edgecolor ='b',label= 'popular')
plt.bar(x,z, width = 0.7, color = 'r', edgecolor ='b',label= 'popular1')

plt.legend() # when label property are need to apply , lenged are used to display this content on bar chart 
plt.show()
```

```python
import numpy as np

x = ["Python" , "C" ,"CPP", "Java"]
y = [3,7,5,4]
z = [2,3,3,2]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p] 

# Labeling to the plot 
plt.xlabel("Programming Language", fontsize = 10)
plt.ylabel("Popular", fontsize = 10)
plt.title("Treading Language",fontsize = 14)


# Diplay the Bar Plot
plt.bar(p,y, width, color = 'pink', edgecolor ='b',label= 'popular')
plt.bar(p1,z, width, color = 'r', edgecolor ='b',label= 'popular1')

plt.xticks(p+width / 2, x , rotation = 35)
plt.legend() # when label property are need to apply , lenged are used to display this content on bar chart 
plt.show()
```

# Scatter Plot 

```python
day = [1,2,3,4,5,6,7,8]
no = [2,3,4,4,3,4,5,4]
color = ['r','b','g','y','r','g','b','y'] #Instead you can use color code
scatter_point_size = [400,200,400,700,300,100,500,300]

plt.title('Scatter Plot',fontsize = 15)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('No', fontsize = 15)

plt.scatter(day,no,c=color,s = scatter_point_size, marker = "*", linewidth = 1.6 , edgecolor = 'pink')
plt.show()
```

# Pie Chart 

```python
x = [10,20,30,40,50]
y = ['C','CPP','Java','Python','.NET']
ex = [0.5,0.0,0.0,0.0,0.0]
c = ['r','g','b','g','y']

plt.pie(x,labels = y , explode = ex , colors = c , autopct = "%0.1f%%" , shadow = True , radius = 0.6 ,
        labeldistance =1.1 , startangle = 90,textprops = {"fontsize":7},counterclock = False , 
        wedgeprops = {'linewidth':5 , 'edgecolor':'m' },rotatelabels = True)

plt.title('Programming Language')
plt.legend(loc = 2)
plt.show()
```

# stem Plot

```python
x = [10,20,30,40,50]
x1 = [50,40,30,20,10]

plt.stem(x,x1,linefmt = ':',markerfmt = 'r>',bottom = 0, basefmt = 'green',label = 'Stem Plot', orientation = 'horizontal') 
# bottom :- starting point
#basefmt :- bottom line color

plt.legend()
plt.show()
```

# Box Plot

```python
x = [10,20,30,40,50,60,70,130]
x1 = [10,20,30,40,50,60,70]
y=[x,x1]
plt.boxplot(y,vert =True,labels = ['python','cpp'],patch_artist = 'True',showmeans = 'True',boxprops =dict(color = 'red'),
           capprops=dict(color='blue'),whiskerprops=dict(color='green')) 
#vert :- vertical False/True (Horizontal/Vertical)
# patch_artist :- fill color into box plot
# showmeans :- determine mean

plt.show()


plt.boxplot(x1 , notch = True) #notch ;- shape of box plot
plt.show()
```

# Stack Plot

```python
x = [1,2,3,4,5,6]
area = [2,4,7,6,3,4]
area1 = [4,3,1,2,4,6]
area2 = [3,5,2,1,2,4]

li = ["Area1","Area2","Area3"]

plt.stackplot(x,area,area1,area2,labels = li , colors = ['r','g','b'])

plt.title('Python')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc = 2)
plt.show()
```

# Step Plot

```python
x = [1,2,3,4,5]
x1 = [2,3,4,5,6]

plt.step(x,x1,color = 'r',marker = '*',ms = 14,mfc='blue',label = 'Python')

plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend()
plt.show()
```

# Fill_Between Plot

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color = 'red')

plt.fill_between(x,y,color = 'blue',where = (x>2) & (x<=4) , alpha = 0.3)

plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

# Text in Matplotlib Plots

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y,label = 'python')

plt.title("Text Plots",fontsize = 12)
plt.xlabel('Days',fontsize = 12)
plt.ylabel('Python',fontsize = 12)

plt.text(2,5,'Java',fontsize = 12,style = 'italic',bbox = ({"facecolor":'pink'}))
plt.annotate("python",xy=(3,1),xytext=(5,5),arrowprops = dict(facecolor = 'black',shrink = 100))

plt.legend(loc = 0,facecolor = 'yellow', edgecolor = 'red',framealpha = 0.6,shadow = True)
plt.show()
```
