# Group By in Pandas

```python
data_ed = pd.read_excel("/Users/ratnadeepgurav/Desktop/ESD.xlsx")

print("\n\t\t <------- Count of Male and Female Department Wise -------> \n\n")

group_by = data_ed.groupby('Department').agg({"Gender":"count"})
print(group_by)

print("\n\t\t <------- Count of Employee ID  Job Title and Gender Wise  -------> \n\n")

group_by = data_ed.groupby(['Job Title','Gender']).agg({"EEID":"count"})
print(group_by)

print("\n\t\t <------- Count of Max Salary and min Age Country Wise  -------> \n\n")

group_by = data_ed.groupby(['Country']).agg({"Annual Salary":"max",'Age':'min'}) # aggregate can max,min,mean,count etc
print(group_by)
```

# Merge , Join And Concatenate In Pandas

```python

data1 = {"EmpId":[101,102,103,104],
         "Name":["John","Saye","Laour","doe"],
         "Age":[55,66,77,88]
        }

data2 = { "EmpId":[101,102,103,104,105],
          "Salary":[22000,22000,66000,77000,88000]
        }

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("\n\t\t <------- Merging Two Dictionary  -------> \n\n")

print(pd.merge(left = df1,right = df2, on = "EmpId", how = 'right')) 

# left column = df1 and right column = df2 and on basis of Common EmpId how = right join
```

# Pandas Compare DataFrame 

```python
import pandas as pd

data_emp1 = {
    "EmpId": [101, 102, 103, 104, 105],
    "Name": ["John", "Saye", "Laour", "Doe", "Sam"],
    "Age": [55, 66, 77, 88, 99]
}

# Create the original DataFrame
data_df1 = pd.DataFrame(data_emp1)

# Make a copy of the original DataFrame
data_df2 = data_df1.copy()

# Add 'Salary' column to the copy
data_df2.loc[0, 'Salary'] = 22000
data_df2.loc[1, 'Salary'] = 23000
data_df2.loc[2, 'Salary'] = 52000
data_df2.loc[3, 'Salary'] = 72000
data_df2.loc[4, 'Salary'] = 55000

# Modify 'Age' column in the copy
data_df2.loc[1, 'Age'] = 35
data_df2.loc[4, 'Age'] = 44

# Ensure both DataFrames have the same columns by adding 'Salary' to data_df1
data_df1['Salary'] = pd.NA  # Add missing 'Salary' column to the original DataFrame

# Now you can safely compare the two DataFrames
print(data_df1.compare(data_df2, keep_equal = True))
```

# Pivoting and Melting DataFrames

```python

# Pivote Table

data = {
    "Keys": ["k1", "k2", "k1", "k2"],
    "Names": ["john", "skyo", "jiha", "nexa"],
    "Houses": ["red", "blue", "green", "red"],
    "Grades":["8th","9th","10th","7th"]
}

df = pd.DataFrame(data)

print(df, "\n\n")

pivot_df = df.pivot(index="Keys", columns="Names", values=["Houses","Grades"])

print(pivot_df)


# Melting DataFrames 


data1 = {
    "Names": ["john", "skyo", "jiha", "nexa"],
    "Houses": ["red", "blue", "green", "red"],
    "Grades": ["8th", "9th", "10th", "7th"]
}
df1 = pd.DataFrame(data1)

# Print the original DataFrame
print(df1, "\n\n")

# Correct melt with var_name and value_name
melting_df1 = df1.melt(id_vars=["Names"], value_vars=["Houses", "Grades"], var_name="Houses&Grades", value_name="values")

# Print the melted DataFrame
print(melting_df1)
```
